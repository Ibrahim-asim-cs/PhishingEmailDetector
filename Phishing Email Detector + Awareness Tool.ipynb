```python
print("Welcome to Phishing Email Detector!")
email_text = input("Paste the email content here: ").lower()

suspicious_words = ["urgent", "verify", "account locked", "click now", "password", "login"]
score = 0

for word in suspicious_words:
    if word in email_text:
        score += 1

if score == 0:
    print("✅ Low risk: Email looks safe.")
elif score <= 2:
    print("⚠️ Medium risk: Be cautious!")
else:
    print("❌ High risk: Potential phishing email!")
```
